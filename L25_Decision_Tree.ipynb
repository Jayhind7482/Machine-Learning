# Decision Tree 

```python
import numpy as np
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
X = data['data']
y=data['target']
feature_name = data['feature_names']
target_name = data['target_names']   # ['setosa', 'versicolor', 'virginica'] [0 ,1,2]
```

```python
df = pd.DataFrame(X , columns=feature_name)
df['target'] = y
```

```python
df.head()
```

```python
X_new = df[['sepal length (cm)' , 'petal length (cm)']]
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X[:,0] , X[:,1] , c=y )
plt.show()
```

```python
clf = DecisionTreeClassifier()
# max_depth = depth of tree , if by default chosen then internally it will split up to when 
# all points are correctly classsified , chances of overfitting (so take 5 ,6)
# criterion = 'gini' or 'entropy' , by default it choses gini , because it is faster then entropy
# min_sample_split = , is the value at each node , if any node has the value less than to this
# set value then further splitting will not happen for that node 
# high value of this hyperparameter is reduce overfitting 
# max_leaf_nodes = is the no max leaf at the end of tree which will decide the class of input
# These are the imp hyperparameter which are important , many more are there 
# if we did not pass anything then it will take the by default value set
```

```python
clf.fit(X_new , y)
```

```python
def plot (X_new , y ,max_depth=None):
    clf2 = DecisionTreeClassifier(max_depth=max_depth)
    clf2.fit(X_new , y)
    X_new = X_new.values
    x_min, x_max = X_new[:, 0].min() - 1, X_new[:, 0].max() + 1
    y_min, y_max = X_new[:, 1].min() - 1, X_new[:, 1].max() + 1
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300)
    )

    # 4. Predict on mesh grid
    Z = clf2.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # 5. Plot
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X_new[:, 0], X_new[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors="k")
    plt.title("Decision Tree Decision Boundary")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
        
```

```python
plot(X_new , y ,max_depth=1)    # Under Fitting
```

```python
plot(X_new , y ,max_depth=2)     # Underfitting 
```

```python
plot(X_new , y ,max_depth=3)   # perfect 
```

```python
plot(X_new , y ,max_depth=10)  # Overfitting
```

```python
plot(X_new , y ) 
# Overfitting , by default max depth is the value till all training insatances is 
# correctly classified 
```
