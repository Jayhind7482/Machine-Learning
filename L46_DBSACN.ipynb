```python
import numpy as np
import pandas as pd 
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
X, y  = make_circles(n_samples=1000 , factor=0.4 , noise=0.1 , random_state=4)
```

```python
X
```

```python
plt.scatter(X[:,0] , X[:,1])
plt.show()
```

```python
dbs = DBSCAN(eps=0.1 , min_samples=5)  
# here eps is radius of circle which will be drawn considering point is centre

# then min_smaples is the minimum sample should be present inside that circle 
# then that center point is called core 
# any point which has less than min_samples inside the circle but it contains atleast one  core point
# that point is called border points 
# and  any point , if a circle is drawn considering the point as centre , if it contains less than 
# min_samples points and does not have any core point then that point is considerd as noise 
# only core point and border point will make cluster , the noise will not be part of any clsuter

# if a new point comes then it cant precit in which cluster it belongs to , same like agglomerative clustering
# hence no predict method 

# this is very powerfull clustering algorithm 

# and here we need not to specify the no of cluster , it will automatically do the clsutering on the basis 
# of density of points 

# for noise it gives cluster number = -1 
```

```python
labels = dbs.fit_predict(X)
```

```python
labels
```

```python
plt.scatter(X[:,0] , X[:,1] , c=labels)
plt.title("DBSCAN Clustering of Concentric Circles")
plt.show()
```
