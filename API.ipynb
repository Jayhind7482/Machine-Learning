```python
import requests
import pandas as pd
import time

API_KEY = "966a60199bmsh6d95d2a2756c043p17c093jsn7687248be8d9"
HOST = "youtube138.p.rapidapi.com"

def search_youtube(query, max_pages=3):
    url = "https://youtube138.p.rapidapi.com/search/"
    headers = {
        "X-RapidAPI-Key": API_KEY,
        "X-RapidAPI-Host": HOST
    }
    
    all_results = []

    for page in range(1, max_pages + 1):
        params = {"q": query, "hl": "en", "gl": "US"}
        response = requests.get(url, headers=headers, params=params)

        if response.status_code != 200:
            print(f"Error: {response.status_code}")
            break

        data = response.json()
        if "contents" not in data:
            break

        for item in data["contents"]:
            if item.get("video"):
                video = item["video"]
                all_results.append({
                    "title": video.get("title", ""),
                    "channel": video.get("author", {}).get("title", ""),
                    "published": video.get("publishedTimeText", ""),
                    "views": video.get("stats", {}).get("views", ""),
                    "videoId": video.get("videoId", ""),
                    "link": f"https://www.youtube.com/watch?v={video.get('videoId', '')}"
                })

        # Optional: Pause to respect rate limits
        time.sleep(1)

    return pd.DataFrame(all_results)


# Example usage
df = search_youtube("cricket highlights", max_pages=2)
#print(df.head())

# Save to CSV
#df.to_csv("youtube_dataset.csv", index=False)

```
