```python
import numpy as np
import pandas as pd 
from sklearn.cluster import KMeans
```

# K-Means Clustering and Elbow Graph

```python
df = pd.read_csv("student_clustering.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.values
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0] , X[:,1])
plt.show()
```

```python
wcss = []
for i in range(1,11):  # considering max 11 clsuter
    kmc = KMeans(n_clusters=i)
    kmc.fit(X)
    y = kmc.predict(X)   # will give each points in X belongs to which clsuter 
    error = kmc.inertia_   # it will give wcss , for 2 clsuter (wcss1**2 + wcss2**2) 
    wcss.append(error)
    
```

```python
plt.plot(range(1,11) ,wcss)
plt.xlabel("K (no of cluster)")
plt.ylabel("WCSS")
plt.show()
```

```python
# from elbow diagram 4 is the optimum no of clsuter 
kmc = KMeans(n_clusters=4)
y_pred = kmc.fit_predict(X)  # will fit and predict the cluster no for each X 
```

K-Means on 3D data

```python
from sklearn.datasets import make_blobs

# Preparing the synthetic data

centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]   # centroid of all 4 cluster 
cluster_std = [1,1,1,1]  # standard daviation of all 4 cluster 

X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

```python
X[1:10 , :]  # Shwoing 10 data points in 3d
```

```python
import plotly.express as px
fig = px.scatter_3d(x=X[:,0] , y=X[:,1] , z=X[:,2] )
fig.show()
```

```python
wcss = []
for i in range(1,21):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,21),wcss)
```

```python
# no of optimum cluster = 4
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
```

```python
df = pd.DataFrame()
df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = y_pred
```

```python
fig = px.scatter_3d(df , x = 'col1' , y ='col2' , z='col3' , color='label')
fig.show()
```
